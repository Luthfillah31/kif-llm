```python
import sys
import os

# 1. Look one folder up, and then into the 'lib' folder
# Since you are in 'examples/', this safely grabs the true source code
lib_path = os.path.abspath(os.path.join('..', 'lib'))

if lib_path not in sys.path:
    sys.path.insert(0, lib_path)

# 2. Import everything natively
from kifqa import KIFQA
from kif_lib import Store, Search

# 3. Initialize the pipeline
kif_wiki_kbqa = KIFQA(
    store=Store('wikidata'),
    search=Search('wikidata-wapi'),
    model_provider='ollama', 
    model_name='deepseek-v3.1:671b-cloud', # Make sure this matches your local model
    model_endpoint='http://localhost:11434', 
    model_apikey='dummy_key', 
    model_params={
        'temperature': 0.0
    }
)

question = 'What is the nationality of anthony bailey'
print(f"Asking: {question}...\n")

# Use .query() based on the authors' demo!
stmts = kif_wiki_kbqa.query(question)

print("Results:")
for stmt in stmts:
    print(stmt)
```
